```python
from nba_api.live.nba.endpoints import scoreboard
from nba_api.live.nba.endpoints import boxscore

# Today's Score Board
games = scoreboard.ScoreBoard()

# json
juegos = games.get_dict()
# dictionary
# games.get_dict()
dict(juegos)
```

```python
game_ids = [game['gameId'] for game in juegos['scoreboard']['games']]
#print(game_ids)

for id in game_ids:
    print(id)
    Stats = boxscore.BoxScore()
    Stats.get_json()
```
